### Calculating daily returns

We will use the formula $r_t = P_t / P_{t-1}-1 = \frac{P_t -P_{t-1}}{P_{t-1}}.$ This is interpretable as "the price rose by x% compared to the previous day". To calculate accumulated wealth: $(1+r_1)(1+r_2)\cdots -1$

```python
# Daily returns
simple_returns = price.pct_change().dropna()
simple_returns.head()
```

### Calculating Logarithmic returns
This are often used for mathematical/statistical analysis. The formula is:

$ l_t = ln(\dfrac{P_t}{P_{t-1}}) $

Log-returns are additive in time, which means the sum of log-returns equals the log of composed return.



```python
# Logarithmic returns
log_returns = np.log(price / price.shift(1)).dropna()  # ln(P_t / P_{t-1})
log_returns.head()
```

### Cummulative return

We will check the accumulated value of an inversion of $1.

```python
# --- Cumulative returns (Growth of $1) ---
cum_simple = (1 + simple_returns).cumprod()  # by columns, evolution of $1 -> cummulative value
cum_price_index = price / price.iloc[0]      # Another way; index of normalized price (P_t / P_0)
cum_price_index.head()
```

Another way of calculating it is using the mathematical property. We will plot the results to observe the growth factor.

```python
cumulative_returns = np.exp(log_returns.cumsum())

cumulative_returns.plot(figsize=(10,5), title="Cumulative Log Returns (Growth of $1)")
plt.xlabel("Date")
plt.ylabel("Growth Factor")
plt.grid(True)
plt.show()
```

### Interpretation

The curves indicate how an investment of $1 in 2005 would have evolved. The shapes of the curves represents compounding effects and volatility drag. 

Comparing the curves, it is shown that on the long run, IVV and GLD performed about the same, showing a similar behaviour in the last decade.

IEF, on the contrary, performed an steadily growth, several times smaller than the other two.

### Correlation and Covariance

Now we will study the underlying diversification potential and correlations between the log returns.

```python
corr = log_returns.corr()
cov = log_returns.cov()

print("Correlation matrix:")
print(corr)

print("\nCovariance matrix:")
print(cov)
```
